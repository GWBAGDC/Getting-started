# Find_metadata.ipynb

This notebook demonstrates how to find out some metadata about the datasets held within the AGDC.

```python
import datacube
```

```python
dc = datacube.Datacube(app='dc-metadata')
```

```python
dc.list_products()
```

```python
dc.list_measurements()
```

## Now we can investigate one dataset in more detail

```python
# Choose a dataset to investigate
dataset = dc.index.products.get_by_name('modis_mcd43a1_tile')
```

```python
# Get some information about the data itself
dataset.measurements
```

```python
# Get some basic metadata
dataset.metadata_doc
```

```python
dataset.definition
```
